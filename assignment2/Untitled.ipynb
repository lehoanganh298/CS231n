```python
import numpy as np

def f(x):
    return x**2-2
def f_prime(x):
    return 2*x

x=100
root=np.sqrt(2)
print(root)
print()
for i in range(10):
    x-=f(x)/f_prime(x)
    print(x, root-x)
```
